### Data Cleaning, EDA, and initial preprocessing on main property transaction sales data from JPPH

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/kuala_lumpur_data.csv")
data.head()
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(r'[ ,/]', '_', regex=True)
data.rename(columns={'month__year_of_transaction_date': 'month_year'}, inplace=True)
data['month_year'] = pd.to_datetime(data['month_year'], format='%B %Y')
data.columns
```

```python
data['year'] = data['month_year'].dt.year
data.drop(columns='month_year', inplace=True)
```
